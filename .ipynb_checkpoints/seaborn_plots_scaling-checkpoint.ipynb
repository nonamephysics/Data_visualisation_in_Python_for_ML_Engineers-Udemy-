```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
sns.set_context('paper')
```

```python
iris = sns.load_dataset('iris')
sns.swarmplot(x = 'species', y = 'petal_length', data = iris)

plt.show()
```

```python
sns.set()
sns.set_context('notebook')

iris = sns.load_dataset('iris')
sns.swarmplot(x = 'species', y = 'petal_length', data = iris)

plt.show()
```

```python
sns.set()
sns.set_context('talk')

iris = sns.load_dataset('iris')
sns.swarmplot(x = 'species', y = 'petal_length', data = iris)

plt.show()
```
