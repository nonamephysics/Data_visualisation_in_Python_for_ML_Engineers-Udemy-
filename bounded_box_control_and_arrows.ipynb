```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = np.linspace(-4, 4, 1024) 
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

box = {'facecolor': '.75', 'edgecolor':'k', 'boxstyle':'round'}
```

```python
plt.annotate ('Low point', ha = 'center', va = 'bottom', xytext = (-1.5, 3.), xy = (.75, -2.7), 
             arrowprops = {'facecolor':'black', 'shrink':0.05})
plt.text(-4, 10, 'Nice Curve in Box', bbox = box)
plt.plot(X, Y, c = 'k') 

plt.show() 
```
