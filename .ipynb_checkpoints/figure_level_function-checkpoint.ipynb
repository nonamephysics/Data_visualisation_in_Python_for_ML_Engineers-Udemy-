```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
titanic = sns.load_dataset("titanic")
```

```python
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", size=7, aspect=2, palette="muted", legend=False)

plt.show()
```

```python
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", size=2, aspect=7, palette="muted", legend=False)

plt.show()
```

```python
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", size=5, aspect=5, palette="muted", legend=False)

plt.show()
```
