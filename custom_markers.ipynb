```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib
```

```python
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30
s = np.random.rand(*x.shape) * 800 + 500
```

```python
plt.scatter(x, y, s, c='g', alpha = .5, marker = r'$\clubsuit$', label = "Luck")
plt.xlabel ('Leprechauns')
plt.ylabel ('Gold')
plt.legend (loc=2)

plt.show()
```
