```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.standard_normal((200, 2))
a += np.array ((-1, -1))
```

```python
b = np.random.standard_normal((200, 2))
b += np.array ((1, 1))
```

```python
plt.scatter(a[:,0], a[:,1], color = '.25')
plt.scatter(b[:,0], b[:,1], color = '.75')

plt.show()
```
