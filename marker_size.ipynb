```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
A = np.random.standard_normal((100, 2)) 
A += np.array((-1, -1))
```

```python
B = np.random.standard_normal((100, 2)) 
B += np.array((1, 1))
```

```python
plt.scatter(B[:,0], B[:,1], c = 'r', marker = 'x', s = 100) 
plt.scatter(A[:,0], A[:,1], c = 'b', marker = '^', s = 50)
plt.show()
```
