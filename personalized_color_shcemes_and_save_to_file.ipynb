```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rc('lines', linewidth = 2.)
mpl.rc('axes', facecolor = 'k', edgecolor = 'w')
mpl.rc('xtick', color = 'w')
mpl.rc('ytick', color = 'c')
mpl.rc('text', color = 'b')
mpl.rc('figure', facecolor = 'k', edgecolor = 'r')
```

```python
x = np.linspace (0, 7, 1024)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.savefig('sin-cos.png', c='k')
```
